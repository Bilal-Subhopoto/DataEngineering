```python
# from tkinter import E
from pymongo import MongoClient
# from datetime import datetime

myclient = MongoClient("mongodb://43.251.253.107:1600/")
#source collection 
sdb = myclient['FB_scraper_MT2']
scol = sdb['Allposts']
```

```python
same_day=[]
one_day_after=[]
two_days_after=[]
three_days_after=[]
four_days_after=[]
five_days_after=[]
six_days_after=[]
seven_days_after=[]
for doc in scol.find({}):
  doc["time"].date()
  doc["TimeDownload"].date()
  delta = doc["TimeDownload"].date()- doc["time"].date()
  
  if delta.days == 0:
    same_day.append(doc["pageName"])
  if delta.days == 1:
    one_day_after.append(doc["pageName"])
  if delta.days == 2:
    two_days_after.append(doc["pageName"])
  if delta.days == 3:
    three_days_after.append(doc["pageName"])
  if delta.days == 4:
    four_days_after.append(doc["pageName"])
  if delta.days == 5:
    five_days_after.append(doc["pageName"])
  if delta.days == 6:
    six_days_after.append(doc["pageName"])
  if delta.days == 7:
    seven_days_after.append(doc["pageName"])

print(same_day)
print(len(same_day) ," Docs with time.date() = TimeDownload.date()")
print(one_day_after)
print(len(one_day_after) ," Docs with time.date() = TimeDownload.date()-1")
print(two_days_after)
print(len(two_days_after) ," Docs with time.date() = TimeDownload.date()-2")
print(three_days_after)
print(len(three_days_after) ," Docs with time.date() = TimeDownload.date()-3")
print(four_days_after)
print(len(four_days_after) ," Docs with time.date() = TimeDownload.date()-4")
print(five_days_after)
print(len(five_days_after) ," Docs with time.date() = TimeDownload.date()-5")
print(six_days_after)
print(len(six_days_after) ," Docs with time.date() = TimeDownload.date()-6")
print(seven_days_after)
print(len(seven_days_after) ," Docs with time.date() = TimeDownload.date()-7")
```
